### Importing All required libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

## Identifying and handling the missing values

```python
df=pd.read_csv(r"..\Mortgage-Prepayment-Analysis-and-Prediction\MortgageLoanExport\LoanExport\LoanExport.csv")
df.head()
```

```python
print(f'The number of rows: {df.shape[0]} \nThe number of columns : {df.shape[1]}')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
type(df['SellerName'][0])
```

#### Here we can see that only one column(SellerName )  has null values and it is a String datatype and so we will missing values with 'unknown'

```python
df['SellerName'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['FirstPaymentDate'] = pd.to_datetime(df['FirstPaymentDate'], format='%Y%m')
df['MaturityDate']=pd.to_datetime(df['MaturityDate'],format='%Y%m')
```

```python
df['NumBorrowers'] = pd.to_numeric(df['NumBorrowers'], errors='coerce')
df['PostalCode'] = pd.to_numeric(df['PostalCode'], errors='coerce')
```

```python
df['NumBorrowers'].fillna(df['NumBorrowers'].mean(),inplace=True)
```

```python
df = df.dropna(subset=['PostalCode'])
```

```python
df.shape
```

```python
df['MSA']=df['MSA'].replace('X', 0)
df['MSA']=pd.to_numeric(df['MSA'], errors='coerce')
```

```python
df.fillna(0,inplace=True)
```

```python
# Step 2: Remove duplicates
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.shape #no duplicates persent
```

```python
df_cleaned.head()
```

```python
target=df_cleaned['EverDelinquent'].values
target
```

```python
data=df_cleaned.drop(['EverDelinquent','LoanSeqNum'],axis=1)
data.shape
```

```python
categorical_columns=['Occupancy','Channel','PPM','ProductType','PropertyState','PropertyType','LoanPurpose','SellerName','ServicerName']
label_column=['FirstTimeHomebuyer']
```

```python
for i in label_column:
    data[i]=df[i].map({'Y': 1, 'N': 0,'X':2})
```

```python
data_encoded=pd.get_dummies(data, columns=categorical_columns, drop_first=True)
data_encoded.head()
```

```python
object_columns = data_encoded.select_dtypes(include=['object']).columns
object_columns
```

```python
df_final = data_encoded.copy()
summary = {
    "Data Types After Correction": df_final.dtypes,
    "Shape After Removing Duplicates": df_cleaned.shape,
    "Final Shape After Encoding": df_final.shape
}

summary
```

```python
data_encoded.head()
```

```python
x=data_encoded.drop(['FirstPaymentDate','MaturityDate','PostalCode'],axis=1)
```

```python
x.columns[x.isna().any()].tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, target, test_size=0.3, random_state=42)

# Initialize and train the classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, target, test_size=0.3, random_state=42)

# Initialize and train the logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Make predictions
y_pred = log_reg.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```
